## 5️⃣ Zip Checkpoint để Download
> Sau khi train xong, chạy cell này để nén checkpoint → download từ tab **Output** của Kaggle.

```python
import shutil, os

CKPT_DIR = 'model_checkpoint'
ZIP_OUT  = '/kaggle/working/model_best'

if os.path.exists(CKPT_DIR):
    shutil.make_archive(ZIP_OUT, 'zip', CKPT_DIR)
    size_mb = os.path.getsize(ZIP_OUT + '.zip') / 1024**2
    print(f'✅ Đã nén: {ZIP_OUT}.zip ({size_mb:.0f} MB)')
    print('👉 Download từ tab Output bên phải của Kaggle!')
else:
    print(f'⚠️  Không tìm thấy thư mục {CKPT_DIR}. Train xong chưa?')
```
